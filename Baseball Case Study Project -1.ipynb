#### Spliting the data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

we are removing those columns which are less contributing to predict out outpyt data.However i checked without removing 
columns but score is much better too exclude it.

```python
X = df_ball.drop(columns=['wins','Earned Run Average','Doubles','At_Bats','Earned_Runs'],axis=1)
y = df_ball['wins']
```

```python
print("Shape of X", X.shape)
print("Shape of Y", y.shape)
```

```python
# Scaling the features using standardscaler
scaler = StandardScaler()
```

```python
x_scale =  scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test  = train_test_split(x_scale, y, test_size=0.25,random_state=42)
```

## Model Building....!

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.svm import SVR
```

```python
lr=LinearRegression()
knn=KNeighborsRegressor()
dt=DecisionTreeRegressor()
rf=RandomForestRegressor()
adb=AdaBoostRegressor()
svm=SVR()
print("Model is created")
```

```python
lr.fit(X_train, y_train)
knn.fit(X_train,y_train)
dt.fit(X_train,y_train)
rf.fit(X_train,y_train)
adb.fit(X_train,y_train)
svm=svm.fit(X_train,y_train)
print("Model is trained")
```

```python
print("lr_score",lr.score(X_train,y_train))
print("knn_score",knn.score(X_train,y_train))
print("dt_score",dt.score(X_train,y_train))
print("rf_score",rf.score(X_train,y_train))
print("adb_score",adb.score(X_train,y_train))
print("svm_score",svm.score(X_train,y_train))
```

```python
print("lr_score",lr.score(X_test,y_test))
print("knn_score",knn.score(X_test,y_test))
print("dt_score",dt.score(X_test,y_test))
print("rf_score",rf.score(X_test,y_test))
print("adb_score",adb.score(X_test,y_test))
print("svm_score",svm.score(X_test,y_test))
```

### Model Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
lr_pred_y = lr.predict(X_test)
knn_pred_y = knn.predict(X_test)
dt_pred_y = dt.predict(X_test)
rf_pred_y = rf.predict(X_test)
adb_pred_y = adb.predict(X_test)
svm_pred_y = svm.predict(X_test)
```

```python
print("lr_score",mean_squared_error(y_test,lr_pred_y))
print("knn_score",mean_squared_error(y_test,knn_pred_y))
print("dt_score",mean_squared_error(y_test,dt_pred_y))
print("rf_score",mean_squared_error(y_test,rf_pred_y))
print("adb_score",mean_squared_error(y_test,adb_pred_y))
print("svm_score",mean_squared_error(y_test,svm_pred_y))
```

```python
print("lr_score",mean_absolute_error(y_test,lr_pred_y))
print("knn_score",mean_absolute_error(y_test,knn_pred_y))
print("dt_score",mean_absolute_error(y_test,dt_pred_y))
print("rf_score",mean_absolute_error(y_test,rf_pred_y))
print("adb_score",mean_absolute_error(y_test,adb_pred_y))
print("svm_score",mean_absolute_error(y_test,svm_pred_y))
```

### HyperParameter Tuning :

```python
from sklearn.model_selection import GridSearchCV
```

```python
lr.get_params().keys()  # to check the parameters
```

```python
param_grid = {'fit_intercept':[True,False],'normalize':[True,False],
              'copy_X':[True,False],'n_jobs':[1,2,3],
              'positive':[True,False]}
param_grid
```

```python
gridsearch = GridSearchCV(lr,param_grid = param_grid,cv=3,verbose = 2 ,n_jobs =4)
```

```python
gridsearch.fit(X_train, y_train)
```

```python
gridsearch.best_params_
```

```python
lr1=LinearRegression(copy_X=True,fit_intercept=True,n_jobs=1,normalize=True,positive=True)
```
